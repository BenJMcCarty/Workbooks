# Correlations

### Graphing Correlations

```python
## Get the correlation matrix for the data (without the target)
corr = df.drop('price',axis=1).corr()
corr.round(2)
```

```python
def multiplot(df,figsize=(15,15),cmap="Reds"):
    
    corr = df.corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    fig, ax = plt.subplots(figsize=figsize)
    sns.heatmap(corr, annot=True,cmap="Reds",mask=mask)
    return fig, ax


multiplot(df.drop('price',axis=1));
```

```python
# Correlation results ignorming (most) duplicate values
df_corr_results = df.corr().unstack().sort_values(ascending=False).drop_duplicates()
df_corr_results
```

```python
# Top 10 most highly correlated results (ignoring one duplicate)
df_corr_results[1:11]
```

```python
# negatively correlated features
df_corr_results[-11:-1]
```
