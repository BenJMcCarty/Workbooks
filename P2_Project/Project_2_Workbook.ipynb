```python
df['date'].iloc[:10]
```

```python
# Select first 7 characters for each element in column

df['date'] = [x[:8] for x in df['date']]

# Convert to datetime
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['date']
```

## Correlations

### Graphing Correlations

Code adapted from https://stackoverflow.com/questions/29432629/plot-correlation-matrix-using-pandas

```python
# Visualizing correlation matrix

f = plt.figure(figsize=(19, 15))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14, rotation=45)
plt.yticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

```python
# Visualizing correlation matrix

corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# Correlation results ignorming (most) duplicate values
df_corr_results = df.corr().unstack().sort_values(ascending=False).drop_duplicates()
df_corr_results
```

```python
# Top 10 most highly correlated results (ignoring one duplicate)
df_corr_results[1:11]
```

```python
# negatively correlated features
df_corr_results[-11:-1]
```
