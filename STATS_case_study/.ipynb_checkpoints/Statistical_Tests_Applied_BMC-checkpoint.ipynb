```python
alc_dict = {888:0, 
            777: np.nan,
            999: np.nan
            }

df['alcohol'] = df['ALCDAY5'].replace(alc_dict)

```

```python
df['alcohol']
```

```python
# create a function called calc monthly
def calc_monthly(x):
    if x>200:
        return x-200
    elif x>100:
        return min(int((x-100)*4.34524), 30)
    else:
        return x
```

```python
df['alcohol']= df['alcohol'].apply(calc_monthly)
```

```python
df['alcohol'].describe()
```
